```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import calendar
import seaborn as sns
```

```python
sales = pd.read_excel('pr_sales.xlsx')
```

```python
sales['new date'] = pd.to_datetime(sales['Year'], format='%Y')
sales['new date'].head()
```

```python
sales.head()
```

```python
sales.drop(59, axis=0, inplace=True)
```

```python
sales['Month'].dtype
```

```python
sales['Month'].apply(lambda x: int(x))
sales['Month'] = sales['Month'].astype(int)
```

```python
sales['Year'] = sales['Year'].apply(lambda x: int(x))
```

```python
sales.head()
```

```python
sales['Month'] = sales['Month'].apply(lambda x: calendar.month_abbr[x])
sales['Month'].dtype
```

```python
sales.head()
```

```python
sales['Year'].apply(lambda x: int(x))
sales['Year'] = sales['Year'].apply(lambda x: "'" + str(abs(x) % 100))
```

```python
sales.head()
```

```python
sales['Date'] = sales['Month'] + ' ' + sales['Year']
```

```python
sales.plot(x="Date", y="Total Sales", kind="bar")
```
